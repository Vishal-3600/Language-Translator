**Import Libraries**

```python
from googletrans import Translator, LANGUAGES
from gtts import gTTS
import speech_recognition as sr
from IPython.display import Audio
```

**Initialize Translator**

```python
translator = Translator()
```

**Show Supported Languages**

```python
print("Supported Languages:\n")

for code, language in LANGUAGES.items():
    print(f"{code} : {language}")
```

**Text Translation Program**

```python
def translate_text():
    text = input("Enter text to translate: ")
    dest_lang = input("Enter destination language code (e.g., hi, fr, es): ")

    translated = translator.translate(text, dest=dest_lang)

    print("\nDetected Language:", translated.src)
    print("Translated Text:", translated.text)

translate_text()
```

**Speech to Text Translation**

```python
def speech_to_text_translate():
    recognizer = sr.Recognizer()

    with sr.Microphone() as source:
        print("Speak something...")
        audio = recognizer.listen(source)

    try:
        text = recognizer.recognize_google(audio)
        print("You said:", text)

        dest_lang = input("Enter destination language code: ")
        translated = translator.translate(text, dest=dest_lang)

        print("Translated Text:", translated.text)

    except Exception as e:
        print("Error:", e)

 #speech_to_text_translate()
```

```python
def text_to_speech(text, lang):
    speech = gTTS(text=text, lang=lang)
    speech.save("translated_audio.mp3")
    return Audio("translated_audio.mp3", autoplay=True)

# Example
translated = translator.translate("Hello Vishal how are you well done bro", dest="hi")
text_to_speech(translated.text, "hi")
```
